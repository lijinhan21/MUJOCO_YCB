### Change the file names correspond to `object name` and add mujoco xml file template

```python
import os
import shutil

original_folder = '../asset/ycb'
new_folder = '../asset/ycb_rename'

# copy the original folder to a new folder
if os.path.exists(new_folder):
    shutil.rmtree(new_folder)
shutil.copytree(original_folder, new_folder)
```

```python
# add mujoco xml files to the new folder
for root, dirs, files in os.walk(new_folder):
    for folder in dirs:
        if (folder != 'tsdf') and (folder != 'google_16k'):
            continue
        
        parent_name = os.path.basename(root)
        obj_name = '_'.join(parent_name.split('_')[1:])
        
        if '-' in parent_name:
          num = ord(parent_name.split('-')[1][0]) - ord('a') + 1
          obj_name = obj_name + '_' + str(num)
        
        print(obj_name)
        
        mujoco_xml = f"""<mujoco model="{obj_name}">
  <asset>
    <mesh file="../obj/{obj_name}.obj" name="{obj_name}_mesh"/>
    <texture file="../png/{obj_name}.png" name="tex-{obj_name}" type="2d"/>
    <material name="mat-{obj_name}" reflectance="0.7" texrepeat="5 5" texture="tex-{obj_name}" texuniform="true"/>
  </asset>
  <worldbody>
    <body>
      <body name="object">
        <geom pos="0 0 0" mesh="{obj_name}_mesh" type="mesh" solimp="0.998 0.998 0.001" solref="0.001 1" density="1000" friction="0.95 0.3 0.1" material="mat-{obj_name}" group="0" condim="4"/>
      </body>
    </body>
  </worldbody>
</mujoco>"""
        
        with open(os.path.join(root, folder, f'{obj_name}.xml'), 'w') as f:
            f.write(mujoco_xml)
        print(f'write {os.path.join(root, folder, f"{obj_name}.xml")}')
```

### Merge: `obj, mtl, png` files into one folder

```python
import os
import shutil

# Set parent directory path
src_folder = '../asset/ycb_rename'
dest_folder = '../asset/ycb_merged'

# Create destination folder (skip if it already exists)
if not os.path.exists(dest_folder):
    os.makedirs(dest_folder)

# Iterate through subfolders according to priority
subfolders_priority = ['google_16k', 'tsdf', 'poisson', 'clouds']

# For each object folder in the parent directory
for object_folder in os.listdir(src_folder):
    object_folder_path = os.path.join(src_folder, object_folder)
    if os.path.isdir(object_folder_path):
        # Iterate through subfolders according to priority
        for subfolder in subfolders_priority:
            subfolder_path = os.path.join(object_folder_path, subfolder)
            if os.path.exists(subfolder_path):
                # Iterate through all files in the subfolder
                for file in os.listdir(subfolder_path):
                    
                    # Absolute path of source file
                    src_file_path = os.path.join(subfolder_path, file)
                    # Absolute path of destination file
                    dest_file_path = os.path.join(dest_folder, file)
                    
                    # If file with same name exists in destination folder, rename it
                    if os.path.exists(dest_file_path):
                        base, extension = os.path.splitext(file)
                        i = 1
                        while os.path.exists(dest_file_path):
                            file = f"{base}_{i}{extension}"
                            dest_file_path = os.path.join(dest_folder, file)
                            i += 1
                    
                    # copy file
                    shutil.copy2(src_file_path, dest_file_path)
                    
                    # For mtl files, modify specific lines
                    if file.endswith('.mtl'):
                        with open(dest_file_path, 'r') as infile:
                            lines = infile.readlines()
                        with open(dest_file_path, 'w') as outfile:
                            for line in lines:
                                if line.startswith('map_Kd'):
                                    line = f'map_Kd {os.path.splitext(file)[0]}.png\n'
                                outfile.write(line)
                    # For obj files, modify specific lines
                    elif file.endswith('.obj'):
                        with open(dest_file_path, 'r') as infile:
                            lines = infile.readlines()
                        with open(dest_file_path, 'w') as outfile:
                            for line in lines:
                                if line.startswith('mtllib'):
                                    line = f'mtllib {os.path.splitext(file)[0]}.mtl\n'
                                outfile.write(line)
                
                # If files were copied from subfolder, move to next object folder
                break

print("All files are merged and mtl/obj files are modified in the folder: ", dest_folder)
```

### Merge v2: Converted mjcf from obj files [xml, obj, png] into each folder

```python
import os
import shutil

src_folder = '../asset/ycb_merged'     # Folder containing original files
dst_folder = '../asset/ycb_merged_v2'  # Folder where files will be moved to
extensions = ['obj', 'png', 'xml']  # List of file extensions

# Create new folder if it doesn't exist
if not os.path.exists(dst_folder):
    os.mkdir(dst_folder)

# Traverse all subdirectories to find files
for dirpath, dirnames, filenames in os.walk(src_folder):
    for filename in filenames:
        # get file extension
        file_extension = os.path.splitext(filename)[1][1:]

        # obj, png, xml extensions
        if file_extension in extensions:
            # Create new folder path (e.g., 'ycb_merged_v2/obj/')
            new_folder_path = os.path.join(dst_folder, file_extension)
            # Create new folder if it doesn't exist
            if not os.path.exists(new_folder_path):
                os.mkdir(new_folder_path)
            
            # Create original file path and new file path
            old_file_path = os.path.join(dirpath, filename)
            new_file_path = os.path.join(new_folder_path, filename)

            # Remove necessary files
            if file_extension == 'xml' and ('nontextured' in filename or 'kinbody' in filename):
                continue
            
            # Move original file to new location
            shutil.copy(old_file_path, new_file_path)
```
